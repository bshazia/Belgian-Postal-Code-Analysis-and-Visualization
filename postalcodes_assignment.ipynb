```python
from pathlib import Path
import re
from geopy.geocoders import Nominatim
import folium
```

```python
postalcode = open('postalcodes.txt','r',encoding='ISO-8859-15')
```

```python
list_codes = postalcode.readlines()
postalcode.close()
```

```python
postal_input = input()
```

```python
anser_list = []

for line in range(0,len(list_codes)):
    count = 0
    for code in str(postal_input):
        anser = code in list_codes[line]
        print(anser,code,list_codes[line])
        if anser == True:
            count += 1
    #print(count)
    if  count == 4:
        anser_list.append(list_codes[line])
```

```python
#print(len(list_codes), len(anser_list))
```

```python
spt_list = []
for x in anser_list:
    spt_list.append(x.split())
    #print(x)
```

```python
#len(spt_list)
```

```python
city_and_Codes_list = ""
result = []
for item in spt_list:
    counter = 0
    for i in item[0]:
        anr = str(i) in str(postal_input)
        if anr == True:
            counter += 1
        if counter == 4:    
            #print(item)
            #result.append(item)
            city_and_Codes_list += ' '.join(item)
            city_and_Codes_list += '\n'
            
    #print(item[0],"hi")
```

```python
print(city_and_Codes_list)
```

```python
def get_coordinates(city):
    geolocator = Nominatim(user_agent="my_geocoder")
    location = geolocator.geocode(city)
    return location.latitude, location.longitude

```

```python
mymap = folium.Map(location=[51.1634, 4.5537], zoom_start=10)
pattern = re.compile(r'(\d+)\s(.+)')
```

```python
for postal_code in city_and_Codes_list:
    match = pattern.match(postal_code)
    if match:
        postal_code_number = match.group(1)  # Extract postal code number
        city = match.group(2)  # Extract city name
        latitude, longitude = get_coordinates(city)
        folium.Marker(
            location=[latitude, longitude],
            popup=f"{postal_code_number}: {city}",
            icon=folium.Icon(color="red", icon="info-sign")
        ).add_to(mymap)
```
